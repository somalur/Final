```python
train_size=int(len(df)*0.9)
test_size=len(df)-train_size
train,test=df.iloc[0:train_size],df.iloc[train_size:len(df)]

print(train.shape,test.shape)
```

```python
from sklearn.preprocessing import RobustScaler
f_columns=['Seven_Day_ADT']

f_transformer=RobustScaler()
adt_transformer=RobustScaler()

f_transformer=f_transformer.fit(train[f_columns].values)
adt_transformer=adt_transformer.fit(train[['Seven_Day_ADT']])


```

```python
def create_dataset(X,y,time_steps=1):
    Xs,Ys=[],[]
    for i in range(len(X)-time_steps):
        v=X.iloc[i:(i+time_steps)].values
        Xs.append(v)
        Ys.append(y.iloc[i+time_steps])
    return np.array(Xs),np.array(Ys)
```

```python
TIME_STEPS=24

X_train, y_train=create_dataset(train,train.Seven_Day_ADT,time_steps=TIME_STEPS)
X_test,y_test=create_dataset(test,test.Seven_Day_ADT,time_steps=TIME_STEPS)
```

```python
print(X_train.shape,y_train.shape)
```

```python
X_test[0][0].shape
```

```python

model=keras.Sequential()
model.add(
    keras.layers.Bidirectional(
        keras.layers.LSTM(
           units=200,
               input_shape=(X_train.shape[1],X_train.shape[2])
                        )
                )
            )


model.add(keras.layers.Dropout(rate=0.5))
model.add(keras.layers.Dense(units=3))
```

```python
#model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
history=model.fit(X_train,y_train,epochs=40,batch_size=200,validation_split=0.4,shuffle=False)
```

```python
model.summary()
```

```python
score, err = model.evaluate(X_test, y_test, verbose=0)
score=int(score)
err=int(err)
print('Test accuracy:', err)
```

```python
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='validation')
plt.legend()
```
